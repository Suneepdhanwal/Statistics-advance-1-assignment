## **THEORY QUESTIONS**

**Ques.1** What is a random variable in probability theory?

**Ans.1** A random variable is a numerical description of the outcome of a random experiment. It can take different values based on the outcome of the random process.

**Ques.2** What are the types of random variables?

**Ans.2** There are two types of random variables:

Discrete Random Variables: These have a countable number of distinct values (e.g., number of students in a class).

Continuous Random Variables: These can take an infinite number of values within a given range (e.g., heights of students).

**Ques.3** What is the difference between discrete and continuous distributions?

**Ans.3** Discrete Distributions: Deal with discrete random variables and list probabilities for each possible outcome.

Continuous Distributions: Deal with continuous random variables and use a probability density function (PDF) to describe the likelihood of different outcomes within a range.

**Ques.4** What are probability distribution functions (PDF)?

**Ans.4** For continuous random variables, a PDF describes the relative likelihood for this random variable to take on a given value. The area under the PDF curve over an interval represents the probability of the random variable falling within that interval.

**Ques.5** How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

**Ans.5** CDF (Cumulative Distribution Function): Shows the probability that a random variable will take a value less than or equal to a specific value.

PDF: Provides the likelihood of the random variable taking on an exact value (for discrete) or within an infinitesimally small interval (for continuous).

**Ques.6** What is a discrete uniform distribution?

**Ans.6** A distribution where all possible outcomes are equally likely. For example, rolling a fair six-sided die.

**Ques.7** What are the key properties of a Bernoulli distribution?

**Ans.7** Represents a random variable with exactly two possible outcomes, "success" and "failure", with a single trial (e.g., flipping a coin).

**Ques.8** What is the binomial distribution, and how is it used in probability?

**Ans.8** Used to model the number of successes in a fixed number of independent Bernoulli trials (e.g., number of heads in 10 coin flips).

**Ques.9** What is the Poisson distribution and where is it applied?

**Ans.9** Describes the probability of a given number of events occurring in a fixed interval of time or space, given the events occur with a known constant mean rate and independently of the time since the last event (e.g., number of emails received in an hour).

**Ques.10** What is a continuous uniform distribution?

**Ans.10** All outcomes are equally likely within a continuous range. For instance, the probability of a point lying in a specific sub-interval of a line segment is proportional to the length of the sub-interval.

**Ques.11** What are the characteristics of a normal distribution?

**Ans.11** A continuous probability distribution characterized by a symmetric, bell-shaped curve centered around the mean, with its shape defined by the mean and standard deviation.

**Ques.12** What is the standard normal distribution, and why is it important?

**Ans.12**A special case of the normal distribution with a mean of 0 and a standard deviation of 1. Used for standardizing scores.

**Ques.13** What is the Central Limit Theorem (CLT), and why is it critical in statistics?

**Ans.13** States that the distribution of the sample mean of a large number of independent, identically distributed variables will be approximately normally distributed, regardless of the underlying distribution.

**Ques.14** How does the Central Limit Theorem relate to the normal distribution?

**Ans.14** CLT explains why the normal distribution arises naturally in many situations and is used as a foundation for various statistical methods.

**Ques.15** What is the application of Z statistics in hypothesis testing?

**Ans.15** Z-statistics measure how many standard deviations an element is from the mean. Used in tests to determine if a sample mean differs significantly from the population mean.

**Ques.16** How do you calculate a Z-score, and what does it represent?

**Ans.16** 𝑍= (𝑋−𝜇) / 𝜎, where, 𝑋  is the value, 𝜇 is the mean, and 𝜎 is the standard deviation. It indicates how many standard deviations a value is from the mean.

**Ques.17** What are point estimates and interval estimates in statistics?

**Ans.17** Point Estimate: A single value estimate of a population parameter (e.g., sample mean).

Interval Estimate: Provides a range within which the parameter is expected to lie (e.g., confidence interval).

**Ques.18** What is the significance of confidence intervals in statistical analysis?

**Ans.18** Confidence intervals provide a range of values that likely contain the population parameter, with a certain level of confidence (e.g., 95% confidence interval).

**Ques.19** What is the relationship between a Z-score and a confidence interval?

**Ans.19** Z-scores determine the critical values for constructing confidence intervals.

**Ques.20** How are Z-scores used to compare different distributions?

**Ans.20** Z-scores standardize different distributions, enabling comparison.

**Ques.21** What are the assumptions for applying the Central Limit Theorem?

**Ans.21** The samples must be independent and identically distributed (i.i.d.), and the sample size should be sufficiently large (usually
n>30).

**Ques.22** What is the concept of expected value in a probability distribution?

**Ans.22** The weighted average of all possible values of a random variable, weighted by their probabilities. Represents the long-term average outcome.

**Ques.23** How does a probability distribution relate to the expected outcome of a random variable?

**Ans.23** The expected value is derived from the probability distribution, indicating the average outcome of a random variable over many trials.

**Ques.9** Write a Python function to calculate Z-scores from a dataset and plot them?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(x - mean) / std for x in data]
    return z_scores

data = np.random.normal(0, 1, 100)
z_scores = calculate_z_scores(data)
sns.histplot(z_scores, bins=20, kde=True)
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.title('Histogram of Z-scores')
plt.show()
```

**Ques.16** Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def standard_normal_pdf(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * x**2)

x = np.linspace(-4, 4, 1000)
y = standard_normal_pdf(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')
plt.show()
```

**Ques.20** Create a confidence interval for a dataset using Python and interpret the result?

```python
import numpy as np
from scipy import stats

data = np.random.normal(0, 1, 100)
mean = np.mean(data)
std = np.std(data)
confidence_level = 0.95
degrees_freedom = len(data) - 1
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, mean, std / np.sqrt(len(data)))

print(f"Confidence interval: {confidence_interval}")
```

**Ques.22** Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 0, 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Normal Distribution')
plt.show()
```

**Ques.23** Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 3
x = np.arange(0, 15, 1)
cdf = poisson.cdf(x, lam)

plt.plot(x, cdf, 'bo', ms=8)
plt.vlines(x, 0, cdf, colors='b', lw=5, alpha=0.5)
plt.xlabel('Number of Events')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of Poisson Distribution')
plt.show()
```

**Ques.25** Write a Python program to compare the standard deviations of two datasets and visualize the difference?

```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(0, 2, 100)

std1 = np.std(data1)
std2 = np.std(data2)

print(f"Standard Deviation of Dataset 1: {std1}")
print(f"Standard Deviation of Dataset 2: {std2}")

plt.hist(data1, bins=20, alpha=0.5, label='Dataset 1')
plt.hist(data2, bins=20, alpha=0.5, label='Dataset 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.title('Comparison of Standard Deviations')
plt.show()
```

**Ques.27** Implement Z-score normalization on a dataset and visualize its transformation?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = np.random.normal(0, 1, 100)
mean = np.mean(data)
std = np.std(data)

z_scores = [(x - mean) / std for x in data]

sns.histplot(z_scores, bins=20, kde=True)
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.title('Z-score Normalization')
plt.show()
```
